```python
import pandas as pd
import numpy as np
```

```python
synopsis = pd.read_csv('anime_with_synopsis.csv')
synopsis
```

```python
# sypnopsis만 추출
content = synopsis['sypnopsis']
content
```

```python
# 토큰생성을 위해 자연어처리 패키지인 nltk패키지를 임포트
import nltk # 형태소 분석 라이브러리
nltk.download('punkt')
```

```python
nltk.download('stopwords')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
print('불용어 개수 :', len(stop_words))
print('불용어 10개 출력 :',stop_words[:30])
```

```python
# 불용어 처리
stopwords = nltk.corpus.stopwords.words('english')

# 검색을 위한 토큰화와 데이터를 소문자처리
text = []
for j in range(11000):
    t1 = nltk.word_tokenize(content[j]) #시놉시스 토큰화 
    for i in range(len(t1)):
         t1[i] = t1[i].lower() # 모든 문자를 소문자화 시킴
    text.append(t1)
```

```python
text
```

```python
def ani_name(search):
    lower = search.lower()
    name = []
    for i in range(11000):
        if lower in text[i]:
            a = synopsis.at[i, 'Name']
            name.append(a)
    return pd.DataFrame(name[:10], columns=["Anime-title"])
```

```python
ani_name(input(prompt = "검색할 시놉시스: "))
```
