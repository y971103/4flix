# 데이터로딩

```python
import numpy as np
import pandas as pd
```

```python
anime = pd.read_excel('anime1.xlsx')
animelist = pd.read_csv('animelist.csv')
```

# 컬럼명 변경

```python
anime.rename(columns = {'MAL_ID':'anime_id'},inplace=True)
```

# 데이터프레임 병합

```python
ani = pd.merge(animelist,anime)
```

```python
#pivot_table을 이용해 유저의 애니메이션 평점을 담은 데이터프레임을 만든다.(title_user)
title_user = ani.pivot_table('rating', index='user_id', columns='Name')
title_user.fillna(0, inplace=True)
title_user
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 유저와 유저 간의 유사도
user_based_collab = cosine_similarity(title_user, title_user)
user_based_collab = pd.DataFrame(user_based_collab, index=title_user.index, columns=title_user.index)
user_based_collab[1995].sort_values(ascending=False)[:10]
```

```python
user_based_collab
```

```python
# 가장 유사한 유저를 뽑아서 해당 유저가 본 영화 추천
user = user_based_collab[1995].sort_values(ascending=False)[:].index[1]
result = title_user.loc[user].sort_values(ascending=False).head()
print(user)
df_r = result.to_frame()
df_r
```
